```python
since_dates = ["2020-01-01", "2020-02-01", "2020-03-01","2020-04-01", "2020-05-01", "2020-06-01", "2020-07-01"]
until_dates = ["2020-01-31", "2020-02-29", "2020-03-31","2020-04-30", "2020-05-31", "2020-06-30", "2020-07-31"]
location = "Philippines"
count = 2
```

```python
df_main = pd.DataFrame(index=None, columns=["ID","Username","AccountDateCreated",
                                                    "TweetDate","Tweet","Retweets",
                                                    "Favorites","Mentions","Hashtags",
                                                    "Location"])
```

```python
for i in zip(since_dates,until_dates):
    
    tweetCriteria = got.manager.TweetCriteria().setTopTweets(True).setNear(location)\
                        .setWithin("200mi").setSince(i[0]).setUntil(i[1])\
                        .setEmoji("unicode").setMaxTweets(count)

    tweets = got.manager.TweetManager.getTweets(tweetCriteria)

    text_tweets = [[tweet.id,tweet.username,api.get_user(str(tweet.username)).created_at,\
                    tweet.date,tweet.text,tweet.retweets,tweet.favorites,tweet.mentions,\
                    tweet.hashtags,api.get_user(str(tweet.username)).location] for tweet in tweets]

    df = pd.DataFrame(text_tweets, index=None, columns=["ID","Username","AccountDateCreated",
                                                        "TweetDate","Tweet","Retweets",
                                                        "Favorites","Mentions","Hashtags",
                                                        "Location"])
    df_main = df_main.append(df, ignore_index = True) 
```
